```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split

# Load and preprocess the MNIST dataset
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

train_set = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_set = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Split train set for calibration and training
train_size = int(0.8 * len(train_set))
calib_size = len(train_set) - train_size
train_set, calib_set = random_split(train_set, [train_size, calib_size])

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
calib_loader = DataLoader(calib_set, batch_size=64, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)
```

```python
import torch.nn as nn
import torch.optim as optim

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = torch.max_pool2d(x, 2)
        x = torch.relu(self.conv2(x))
        x = torch.max_pool2d(x, 2)
        x = x.view(-1, 64 * 7 * 7)
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

# Initialize model, loss function, and optimizer
model = SimpleCNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
for epoch in range(5):
    model.train()
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

```python
def conformal_calibration(model, calib_loader):
    model.eval()
    scores = []
    with torch.no_grad():
        for images, labels in calib_loader:
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)
            max_probs, _ = probs.max(dim=1)
            scores.extend(1 - max_probs)
    # Calculate quantile at desired significance level (e.g., 90% prediction interval)
    epsilon = 0.1  # 10% error tolerance
    q_hat = torch.quantile(torch.tensor(scores), 1 - epsilon).item()
    return q_hat

# Get the calibration quantile for 90% confidence
q_hat = conformal_calibration(model, calib_loader)
print("Calibration Quantile (q_hat):", q_hat)
```

```python
def fgsm_attack(image, epsilon, data_grad):
    sign_data_grad = data_grad.sign()
    perturbed_image = image + epsilon * sign_data_grad
    return torch.clamp(perturbed_image, 0, 1)

# Generate adversarial examples
epsilon = 0.2
model.eval()
adv_examples = []
for images, labels in test_loader:
    images.requires_grad = True
    outputs = model(images)
    loss = criterion(outputs, labels)
    model.zero_grad()
    loss.backward()
    data_grad = images.grad.data
    perturbed_data = fgsm_attack(images, epsilon, data_grad)
    adv_examples.append((perturbed_data, labels))
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import numpy as np

# Data augmentation for training and normalization for test
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

train_size = int(0.8 * len(train_set))
calib_size = len(train_set) - train_size
train_set, calib_set = random_split(train_set, [train_size, calib_size])

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
calib_loader = DataLoader(calib_set, batch_size=64, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)

# Model architecture: ResNet18
model = resnet18(num_classes=10)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
for epoch in range(7):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
      images.requires_grad = True
      optimizer.zero_grad()
      outputs = model(images)
      loss = criterion(outputs, labels)
      loss.backward()
      data_grad = images.grad.data
      adv_images = fgsm_attack(images, epsilon, data_grad)  # Generate adversarial examples
      optimizer.step()

    print(f"Epoch {epoch+1}, Loss: {running_loss / len(train_loader):.4f}")

# Conformal calibration
def conformal_calibration(model, calib_loader, epsilon=0.1):
    model.eval()
    scores = []
    with torch.no_grad():
        for images, labels in calib_loader:
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)
            max_probs, _ = probs.max(dim=1)
            scores.extend(1 - max_probs)
    q_hat = torch.quantile(torch.tensor(scores), 1 - epsilon).item()
    return q_hat

q_hat = conformal_calibration(model, calib_loader, epsilon=0.2)
print("Calibration Quantile (q_hat):", q_hat)

# FGSM Attack
def fgsm_attack(image, epsilon, data_grad):
    perturbed_image = image + epsilon * data_grad.sign()
    return torch.clamp(perturbed_image, -1, 1)

# PGD Attack
def pgd_attack(image, model, labels, epsilon, alpha, num_iter):
    perturbed_image = image.clone()
    for _ in range(num_iter):
        perturbed_image.requires_grad = True
        outputs = model(perturbed_image)
        loss = criterion(outputs, labels)
        model.zero_grad()
        loss.backward()
        grad = perturbed_image.grad.data
        perturbed_image = perturbed_image + alpha * grad.sign()
        perturbed_image = torch.clamp(perturbed_image, image - epsilon, image + epsilon)
        perturbed_image = torch.clamp(perturbed_image, -1, 1)
    return perturbed_image

# Backdoor Attack
def add_backdoor(image):
    image[:, 14:18, 14:18] = 1.0  # Central trigger
    return image

# Apply backdoor to a portion of the training set
backdoor_labels = 0  # Assign backdoor images to class 0
backdoor_percentage = 0.1
num_backdoors = int(len(train_set.dataset) * backdoor_percentage)

for i in range(num_backdoors):
    image, _ = train_set.dataset[i]
    image = add_backdoor(image)
    image = (image - torch.tensor([0.5, 0.5, 0.5]).view(3, 1, 1)) / torch.tensor([0.5, 0.5, 0.5]).view(3, 1, 1)
    train_set.dataset.data[i] = (image.permute(1, 2, 0) * 255).byte().numpy()
    train_set.dataset.targets[i] = backdoor_labels

# Function to compute interval bounds for ReLU
def relu_bounds(lower, upper):
    return torch.relu(lower), torch.relu(upper)

# Forward pass with IBP
def forward_with_ibp(model, x, epsilon):
    x_min, x_max = x - epsilon, x + epsilon

    for layer in model.children():
        if isinstance(layer, nn.Conv2d) or isinstance(layer, nn.Linear):
            x_min = layer(x_min)
            x_max = layer(x_max)
        elif isinstance(layer, nn.ReLU):
            x_min, x_max = relu_bounds(x_min, x_max)
        elif isinstance(layer, nn.MaxPool2d) or isinstance(layer, nn.AvgPool2d):
            x_min = layer(x_min)
            x_max = layer(x_max)
        else:
            raise NotImplementedError(f"IBP not implemented for {type(layer)}")

    return x_min, x_max

# Compute robust loss
def robust_loss(model, x, labels, epsilon):
    x_min, x_max = forward_with_ibp(model, x, epsilon)
    logits_min = torch.log_softmax(x_min, dim=1)
    logits_max = torch.log_softmax(x_max, dim=1)
    robust_logit = torch.min(logits_min, logits_max)
    return criterion(robust_logit, labels)

# Updated training loop with IBP
epsilon = 0.03  # Perturbation bound
for epoch in range(7):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        optimizer.zero_grad()

        # Compute robust loss using IBP
        loss = robust_loss(model, images, labels, epsilon)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {running_loss / len(train_loader):.4f}")

# Evaluate the model using interval bounds
def evaluate_with_ibp(model, test_loader, epsilon):
    model.eval()
    correct = 0
    total = 0
    for images, labels in test_loader:
        with torch.no_grad():
            x_min, x_max = forward_with_ibp(model, images, epsilon)
            logits_min = torch.log_softmax(x_min, dim=1)
            logits_max = torch.log_softmax(x_max, dim=1)
            robust_logit = torch.min(logits_min, logits_max)
            _, predicted = robust_logit.max(1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    accuracy = correct / total
    print(f"Robust Accuracy with epsilon={epsilon}: {accuracy:.4f}")

# Evaluate the model with IBP
evaluate_with_ibp(model, test_loader, epsilon)

# Updated Evaluation function
def evaluate_all_attacks(model, test_loader, q_hat, epsilon, pgd_params=None):
    def evaluate(loader, description, adversarial=None, backdoor=False):
        correct = 0
        total = 0
        prediction_set_sizes = []

        for images, labels in loader:
            if adversarial == "FGSM":
                images.requires_grad = True
                outputs = model(images)
                loss = criterion(outputs, labels)
                model.zero_grad()
                loss.backward()
                data_grad = images.grad.data
                images = fgsm_attack(images, epsilon, data_grad)

            elif adversarial == "PGD":
                assert pgd_params is not None, "PGD parameters must be provided!"
                images = pgd_attack(images, model, labels,
                                    epsilon=pgd_params["epsilon"],
                                    alpha=pgd_params["alpha"],
                                    num_iter=pgd_params["num_iter"])

            if backdoor:
                for i in range(images.size(0)):
                    images[i] = add_backdoor(images[i])

            with torch.no_grad():
                outputs = model(images)
                probs = torch.softmax(outputs, dim=1)
                prediction_sets = (probs >= (1 - q_hat)).float().sum(dim=1)
                prediction_set_sizes.extend(prediction_sets.tolist())
                _, predicted = outputs.max(1)
                correct += (predicted == labels).sum().item()
                total += labels.size(0)

        accuracy = correct / total
        avg_set_size = np.mean(prediction_set_sizes)
        print(f"{description} - Accuracy: {accuracy:.4f}, Avg Prediction Set Size: {avg_set_size:.2f}")

    print("\n--- Evaluating on Clean Test Data ---")
    evaluate(test_loader, "Clean Data")

    print("\n--- Evaluating on FGSM Adversarial Test Data ---")
    evaluate(test_loader, "FGSM Adversarial Data", adversarial="FGSM")

    print("\n--- Evaluating on PGD Adversarial Test Data ---")
    evaluate(test_loader, "PGD Adversarial Data", adversarial="PGD")

    print("\n--- Evaluating on Backdoor Test Data ---")
    evaluate(test_loader, "Backdoor Data", backdoor=True)

# Parameters for PGD attack
pgd_params = {
    "epsilon": 0.3,
    "alpha": 0.01,
    "num_iter": 40
}

epsilon = 0.3
evaluate_all_attacks(model, test_loader, q_hat, epsilon, pgd_params)
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import numpy as np

# ----------------------------
# Data Preparation
# ----------------------------
def get_dataloaders(batch_size=64):
    transform_train = transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.RandomCrop(32, padding=4),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])

    transform_test = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])

    train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
    test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

    train_size = int(0.8 * len(train_set))
    calib_size = len(train_set) - train_size
    train_set, calib_set = random_split(train_set, [train_size, calib_size])

    train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
    calib_loader = DataLoader(calib_set, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_set, batch_size=batch_size, shuffle=False)

    return train_loader, calib_loader, test_loader

train_loader, calib_loader, test_loader = get_dataloaders()

# ----------------------------
# Model, Loss, and Optimizer
# ----------------------------
model = resnet18(num_classes=10)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ----------------------------
# Adversarial Attacks
# ----------------------------
def fgsm_attack(image, epsilon, data_grad):
    perturbed_image = image + epsilon * data_grad.sign()
    return torch.clamp(perturbed_image, -1, 1)

def pgd_attack(image, model, labels, epsilon, alpha, num_iter):
    perturbed_image = image.clone().detach().requires_grad_(True)  # Ensure it's a leaf tensor with grad enabled
    for _ in range(num_iter):
        outputs = model(perturbed_image)
        loss = criterion(outputs, labels)
        model.zero_grad()
        loss.backward()
        grad = perturbed_image.grad.data
        perturbed_image = perturbed_image + alpha * grad.sign()
        perturbed_image = torch.clamp(perturbed_image, image - epsilon, image + epsilon)
        perturbed_image = torch.clamp(perturbed_image, -1, 1)
        perturbed_image = perturbed_image.detach().requires_grad_(True)  # Detach and re-enable grad computation
    return perturbed_image


# ----------------------------
# Conformal Calibration with Adversarial Examples
# ----------------------------
def conformal_calibration_robust(model, loader, epsilon=0.1, attack_type="FGSM"):
    model.eval()
    scores = []
    for images, labels in loader:
        if attack_type == "FGSM":
            images.requires_grad = True  # Enable gradient computation for FGSM
            outputs = model(images)
            loss = criterion(outputs, labels)
            model.zero_grad()
            loss.backward()
            data_grad = images.grad.data
            images = fgsm_attack(images, epsilon, data_grad).detach()  # Detach to make it a leaf tensor
        elif attack_type == "PGD":
            pgd_params = {"epsilon": epsilon, "alpha": 0.01, "num_iter": 10}
            images = pgd_attack(images, model, labels, **pgd_params).detach()

        with torch.no_grad():  # Safeguard for subsequent computations
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)
            max_probs, _ = probs.max(dim=1)
            scores.extend(1 - max_probs)

    return torch.quantile(torch.tensor(scores), 1 - epsilon).item()



q_hat_robust = conformal_calibration_robust(model, calib_loader, epsilon=0.2, attack_type="FGSM")

# ----------------------------
# Robust Evaluation
# ----------------------------
def evaluate_conformal_robust(model, loader, q_hat, epsilon=0.03, attack_type=None):
    model.eval()
    correct, total = 0, 0
    for images, labels in loader:
        if attack_type == "FGSM":
            images.requires_grad = True
            outputs = model(images)
            loss = criterion(outputs, labels)
            model.zero_grad()
            loss.backward()
            data_grad = images.grad.data
            images = fgsm_attack(images, epsilon, data_grad)
        elif attack_type == "PGD":
            pgd_params = {"epsilon": epsilon, "alpha": 0.01, "num_iter": 10}
            images = pgd_attack(images, model, labels, **pgd_params)

        with torch.no_grad():
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)
            max_probs, _ = probs.max(dim=1)
            conformal_set = (1 - probs <= q_hat)
            correct += (conformal_set[range(len(labels)), labels]).sum().item()
            total += labels.size(0)

    print(f"Coverage: {correct / total:.4f}")

# Evaluate robust conformal prediction
print("Robust Conformal Evaluation with FGSM:")
evaluate_conformal_robust(model, test_loader, q_hat_robust, epsilon=0.03, attack_type="FGSM")

print("Robust Conformal Evaluation with PGD:")
evaluate_conformal_robust(model, test_loader, q_hat_robust, epsilon=0.3, attack_type="PGD")
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import numpy as np
import matplotlib.pyplot as plt

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Data augmentation for training and normalization for testing
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load CIFAR-10 dataset
train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

train_size = int(0.8 * len(train_set))
calib_size = len(train_set) - train_size
train_set, calib_set = random_split(train_set, [train_size, calib_size])

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
calib_loader = DataLoader(calib_set, batch_size=64, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)

# Model architecture
model = resnet18(num_classes=10)
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)

# PGD Attack
def pgd_attack(model, images, labels, epsilon, alpha, iters):
    images = images.clone().detach().to(device)
    original_images = images.clone().detach()

    for _ in range(iters):
        images.requires_grad = True
        outputs = model(images)
        loss = criterion(outputs, labels)
        model.zero_grad()
        loss.backward()
        data_grad = images.grad

        # Update images
        adv_images = images + alpha * data_grad.sign()

        # Project perturbation to stay within epsilon
        eta = torch.clamp(adv_images - original_images, min=-epsilon, max=epsilon)
        images = torch.clamp(original_images + eta, min=-1, max=1).detach()

    return images

# Training loop
for epoch in range(1):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch + 1}, Loss: {running_loss / len(train_loader):.4f}")

# RSCP: Conformal Prediction with Randomized Smoothing
def conformal_rscp(model, calib_loader, sigma, ns, epsilon=0.1):
    model.eval()
    scores = []

    with torch.no_grad():
        for images, labels in calib_loader:
            images = images.to(device)
            batch_scores = []

            for _ in range(ns):
                noise = torch.normal(0, sigma, size=images.shape).to(device)
                noisy_images = images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                max_probs, _ = probs.max(dim=1)
                batch_scores.extend((1 - max_probs).cpu().numpy())

            scores.extend(batch_scores)

    q_hat = np.quantile(scores, 1 - epsilon)
    return q_hat

sigma = 0.1
ns = 10
q_hat = conformal_rscp(model, calib_loader, sigma=sigma, ns=ns, epsilon=0.2)
print(f"RSCP Quantile (q_hat): {q_hat:.4f}")

def evaluate_rscp(model, test_loader, q_hat, sigma, ns, epsilon=0.03, alpha=0.01, iters=10):
    model.eval()
    correct_clean = 0
    correct_adv = 0
    total = 0
    conformal_set_sizes = []

    for batch in test_loader:
        # Handle cases where additional data is returned by the DataLoader
        if len(batch) == 2:
            images, labels = batch
        else:
            images, labels, *_ = batch

        images, labels = images.to(device), labels.to(device)

        # Clean evaluation
        with torch.no_grad():
            outputs = model(images)
            _, predicted = outputs.max(1)
            correct_clean += (predicted == labels).sum().item()

            # Conformal prediction set using RSCP
            smooth_probs = []
            for _ in range(ns):
                noise = torch.normal(0, sigma, size=images.shape).to(device)
                noisy_images = images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                smooth_probs.append(probs)

            avg_probs = torch.stack(smooth_probs).mean(dim=0)
            set_sizes = (avg_probs >= q_hat).sum(dim=1).cpu().numpy()
            conformal_set_sizes.extend(set_sizes)

        # PGD adversarial evaluation
        adv_images = pgd_attack(model, images, labels, epsilon, alpha, iters)
        with torch.no_grad():
            adv_outputs = model(adv_images)
            _, predicted_adv = adv_outputs.max(1)
            correct_adv += (predicted_adv == labels).sum().item()

        total += labels.size(0)

    clean_accuracy = correct_clean / total
    adv_accuracy = correct_adv / total
    avg_conformal_set_size = np.mean(conformal_set_sizes)

    return clean_accuracy, adv_accuracy, avg_conformal_set_size

# Generate graph of accuracies
epsilons = np.linspace(0, 0.1, 10)
clean_accuracies = []
adv_accuracies = []
conformal_set_sizes = []

for epsilon in epsilons:
    clean_acc, adv_acc, avg_set_size = evaluate_rscp(model, test_loader, q_hat, sigma, ns, epsilon=epsilon, alpha=0.01, iters=10)
    clean_accuracies.append(clean_acc)
    adv_accuracies.append(adv_acc)
    conformal_set_sizes.append(avg_set_size)
    print(f"Epsilon: {epsilon:.2f} - Clean Acc: {clean_acc:.4f}, Adv Acc: {adv_acc:.4f}, Avg Set Size: {avg_set_size:.2f}")

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(epsilons, clean_accuracies, label="Clean Accuracy")
plt.plot(epsilons, adv_accuracies, label="Adversarial Accuracy (PGD)", linestyle="--")
plt.xlabel("Epsilon (Attack Strength)")
plt.ylabel("Accuracy")
plt.title("Model Accuracy vs. Epsilon (PGD Attack)")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(epsilons, conformal_set_sizes, label="Avg Conformal Set Size (RSCP)", color="purple")
plt.xlabel("Epsilon (Attack Strength)")
plt.ylabel("Average Conformal Set Size")
plt.title("Conformal Set Size vs. Epsilon (RSCP)")
plt.legend()
plt.grid()
plt.show()
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import numpy as np
import matplotlib.pyplot as plt

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Data augmentation for training and normalization for testing
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load CIFAR-10 dataset
train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

train_size = int(0.8 * len(train_set))
calib_size = len(train_set) - train_size
train_set, calib_set = random_split(train_set, [train_size, calib_size])

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
calib_loader = DataLoader(calib_set, batch_size=64, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)

# Model architecture
model = resnet18(num_classes=10)
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)

# PGD Attack
def pgd_attack(model, images, labels, epsilon, alpha, iters):
    images = images.clone().detach().to(device)
    original_images = images.clone().detach()

    for _ in range(iters):
        images.requires_grad = True
        outputs = model(images)
        loss = criterion(outputs, labels)
        model.zero_grad()
        loss.backward()
        data_grad = images.grad

        # Update images
        adv_images = images + alpha * data_grad.sign()

        # Project perturbation to stay within epsilon
        eta = torch.clamp(adv_images - original_images, min=-epsilon, max=epsilon)
        images = torch.clamp(original_images + eta, min=-1, max=1).detach()

    return images

# Training loop
for epoch in range(50):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Generate adversarial examples
        adv_images = pgd_attack(model, images, labels, epsilon=0.03, alpha=0.01, iters=5)

        optimizer.zero_grad()
        # Compute loss on both clean and adversarial inputs
        clean_outputs = model(images)
        adv_outputs = model(adv_images)
        loss = criterion(clean_outputs, labels) + criterion(adv_outputs, labels)
        loss.backward()

        optimizer.step()
        running_loss += loss.item()

    print(f"Epoch {epoch + 1}, Loss: {running_loss / len(train_loader):.4f}")

# RSCP with Adversarial Examples
def conformal_rscp_with_adv(model, calib_loader, sigma, ns, epsilon, alpha, iters, adv_included=True):
    model.eval()
    scores = []

    for images, labels in calib_loader:
        images, labels = images.to(device), labels.to(device)

        # Original scores (clean examples)
        batch_scores = []
        with torch.no_grad():
            for _ in range(ns):
                noise = torch.normal(0, sigma, size=images.shape).to(device)
                noisy_images = images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                max_probs, _ = probs.max(dim=1)
                batch_scores.extend((1 - max_probs).detach().cpu().numpy())

        # Adversarial scores
        if adv_included:
            adv_images = pgd_attack(model, images, labels, epsilon, alpha, iters)
            for _ in range(ns):
                noise = torch.normal(0, sigma, size=adv_images.shape).to(device)
                noisy_images = adv_images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                max_probs, _ = probs.max(dim=1)
                batch_scores.extend((1 - max_probs).detach().cpu().numpy())

        scores.extend(batch_scores)

    # Compute the quantile based on scores
    q_hat = np.quantile(scores, 1 - epsilon)
    return q_hat

# RSCP with Coverage Calculation
def evaluate_rscp_with_coverage(model, test_loader, q_hat, sigma, ns, epsilon, alpha, iters):
    model.eval()
    correct_clean = 0
    correct_adv = 0
    total = 0
    true_labels = []
    conformal_prediction_sets = []
    set_sizes = []

    for batch in test_loader:
        images, labels = batch[:2]
        images, labels = images.to(device), labels.to(device)

        # Save true labels for coverage calculation
        true_labels.extend(labels.cpu().numpy())

        # Clean evaluation
        with torch.no_grad():
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)

            # Generate conformal prediction sets
            prediction_sets = [torch.where(probs[i] >= q_hat)[0].cpu().numpy() for i in range(probs.size(0))]
            conformal_prediction_sets.extend(prediction_sets)
            set_sizes.extend([len(pred_set) for pred_set in prediction_sets])

            _, predicted = outputs.max(1)
            correct_clean += (predicted == labels).sum().item()

        # PGD adversarial evaluation
        adv_images = pgd_attack(model, images, labels, epsilon, alpha, iters)
        with torch.no_grad():
            adv_outputs = model(adv_images)
            _, predicted_adv = adv_outputs.max(1)
            correct_adv += (predicted_adv == labels).sum().item()

        total += labels.size(0)

    # Calculate coverage
    coverage = sum(label in pred_set for label, pred_set in zip(true_labels, conformal_prediction_sets)) / len(true_labels)
    clean_accuracy = correct_clean / total
    adv_accuracy = correct_adv / total
    avg_set_size = np.mean(set_sizes)

    return clean_accuracy, adv_accuracy, coverage, avg_set_size

# Parameters
sigma = 0.1
ns = 20
epsilon = 0.2
alpha = 0.01
iters = 10

# Calculate q_hat
q_hat = conformal_rscp_with_adv(
    model=model,
    calib_loader=calib_loader,
    sigma=sigma,
    ns=ns,
    epsilon=epsilon,
    alpha=alpha,
    iters=iters,
    adv_included=True
)

print(f"RSCP Quantile (q_hat): {q_hat:.4f}")

# Evaluate with multiple epsilon values
epsilons = np.linspace(0, 0.1, 10)
clean_accuracies = []
adv_accuracies = []
coverages = []
set_sizes = []

for epsilon in epsilons:
    clean_acc, adv_acc, coverage, avg_set_size = evaluate_rscp_with_coverage(
        model=model,
        test_loader=test_loader,
        q_hat=q_hat,
        sigma=sigma,
        ns=ns,
        epsilon=epsilon,
        alpha=0.01,
        iters=10
    )
    clean_accuracies.append(clean_acc)
    adv_accuracies.append(adv_acc)
    coverages.append(coverage)
    set_sizes.append(avg_set_size)

    print(f"Epsilon: {epsilon:.2f} - Clean Acc: {clean_acc:.4f}, Adv Acc: {adv_acc:.4f}, Coverage: {coverage:.4f}, Avg Set Size: {avg_set_size:.2f}")

# Plot accuracies
plt.figure(figsize=(10, 6))
plt.plot(epsilons, clean_accuracies, label="Clean Accuracy")
plt.plot(epsilons, adv_accuracies, label="Adversarial Accuracy", linestyle="--")
plt.xlabel("Epsilon (Attack Intensity)")
plt.ylabel("Accuracy")
plt.title("Model Accuracy vs. Attack Intensity")
plt.legend()
plt.grid()
plt.show()

# Plot coverage and set size
plt.figure(figsize=(10, 6))
plt.plot(epsilons, coverages, label="Coverage")
plt.plot(epsilons, set_sizes, label="Avg Conformal Set Size", linestyle="--")
plt.xlabel("Epsilon (Attack Intensity)")
plt.ylabel("Value")
plt.title("Coverage and Set Size vs. Attack Intensity")
plt.legend()
plt.grid()
plt.show()
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import numpy as np
import matplotlib.pyplot as plt

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data augmentation and normalization
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load CIFAR-10 dataset
train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)

# Define a simple trigger (white square in the bottom-right corner)
def add_backdoor_trigger(images, trigger_value=1.0):
    """
    Adds a white square trigger to the bottom-right corner of images.
    Handles both single images (3D) and batched images (4D).
    """
    if images.ndimension() == 4:  # Batched images
        h, w = images.shape[-2:]
        trigger_size = min(5, h, w)  # Ensure trigger fits within the image
        images[:, :, -trigger_size:, -trigger_size:] = trigger_value
    elif images.ndimension() == 3:  # Single image
        h, w = images.shape[-2:]
        trigger_size = min(5, h, w)  # Ensure trigger fits within the image
        images[:, -trigger_size:, -trigger_size:] = trigger_value
    else:
        raise ValueError("Unsupported image dimensions. Expected 3D or 4D tensor.")
    return images


# Inject backdoor samples into the training dataset
def inject_backdoor(data_loader, target_class, poison_fraction=0.1):
    poisoned_images = []
    poisoned_labels = []
    clean_images = []
    clean_labels = []

    for images, labels in data_loader:
        for img, label in zip(images, labels):
            if np.random.rand() < poison_fraction:
                img = add_backdoor_trigger(img)
                poisoned_images.append(img)
                poisoned_labels.append(target_class)
            else:
                clean_images.append(img)
                clean_labels.append(label)

    all_images = torch.stack(clean_images + poisoned_images)
    all_labels = torch.tensor(clean_labels + poisoned_labels, dtype=torch.long)

    return DataLoader(torch.utils.data.TensorDataset(all_images, all_labels), batch_size=64, shuffle=True)

# Initialize a poisoned training loader with a target class
poisoned_train_loader = inject_backdoor(train_loader, target_class=0, poison_fraction=0.1)

# Define the model
model = resnet18(num_classes=10)
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train function
def train(model, loader, optimizer, criterion):
    model.train()
    running_loss = 0.0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    return running_loss / len(loader)
# Function to generate adversarial examples using FGSM
def generate_adversarial_examples(model, images, labels, epsilon=0.03):
    images = images.clone().detach().to(device)
    images.requires_grad = True

    outputs = model(images)
    loss = criterion(outputs, labels)
    model.zero_grad()
    loss.backward()

    perturbation = epsilon * images.grad.sign()
    adversarial_images = torch.clamp(images + perturbation, min=-1, max=1)  # Ensure valid range
    return adversarial_images

# Train function with adversarial examples
def train_with_adversarial_examples(model, loader, optimizer, criterion, epsilon=0.03):
    model.train()
    running_loss = 0.0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        # Generate adversarial examples
        adv_images = generate_adversarial_examples(model, images, labels, epsilon)

        # Combine clean and adversarial images
        combined_images = torch.cat([images, adv_images])
        combined_labels = torch.cat([labels, labels])

        # Forward pass
        outputs = model(combined_images)
        loss = criterion(outputs, combined_labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    return running_loss / len(loader)

# Evaluate function
def evaluate(model, loader, use_backdoor=False):
    model.eval()
    correct = 0
    total = 0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        if use_backdoor:
            images = add_backdoor_trigger(images)

        with torch.no_grad():
            outputs = model(images)
            _, predicted = outputs.max(1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    return correct / total

# RSCP: Conformal Prediction with Randomized Smoothing
def conformal_rscp(model, calibration_loader, sigma, num_samples, epsilon):
    model.eval()
    scores = []

    for images, _ in calibration_loader:
        images = images.to(device)

        # Generate scores for each sample
        with torch.no_grad():
            for _ in range(num_samples):
                noise = torch.normal(0, sigma, size=images.shape).to(device)
                noisy_images = images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                max_probs, _ = probs.max(dim=1)
                scores.extend((1 - max_probs).cpu().numpy())

    # Compute quantile for calibration
    q_hat = np.quantile(scores, 1 - epsilon)
    return q_hat

def evaluate_rscp_with_backdoor(model, loader, q_hat, sigma, num_samples, use_backdoor=False):
    model.eval()
    total = 0
    correct = 0
    covered = 0
    total_set_size = 0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)
        total += labels.size(0)

        if use_backdoor:
            images = add_backdoor_trigger(images)

        with torch.no_grad():
            smooth_probs = []
            for _ in range(num_samples):
                noise = torch.normal(0, sigma, size=images.shape).to(device)
                noisy_images = images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                smooth_probs.append(probs)

            avg_probs = torch.stack(smooth_probs).mean(dim=0)
            prediction_sets = avg_probs >= q_hat

            for i, label in enumerate(labels):
                if prediction_sets[i, label].item():
                    covered += 1
                total_set_size += prediction_sets[i].sum().item()

            _, predicted = avg_probs.max(dim=1)
            correct += (predicted == labels).sum().item()

    accuracy = correct / total
    coverage = covered / total
    avg_set_size = total_set_size / total
    return accuracy, coverage, avg_set_size

# RSCP Parameters
sigma = 0.1
num_samples = 10
epsilon = 0.1

# Create calibration loader
train_size = int(0.8 * len(train_set))
calib_size = len(train_set) - train_size
train_set, calib_set = random_split(train_set, [train_size, calib_size])
calibration_loader = DataLoader(calib_set, batch_size=64, shuffle=True)

# Training loop
# Training loop with adversarial examples
epochs = 50
epsilon = 0.03  # Strength of adversarial perturbations

for epoch in range(epochs):
    train_loss = train_with_adversarial_examples(model, poisoned_train_loader, optimizer, criterion, epsilon=epsilon)
    print(f"Epoch {epoch + 1}/{epochs}, Loss: {train_loss:.4f}")

# Calibrate RSCP
q_hat = conformal_rscp(model, calibration_loader, sigma, num_samples, epsilon)
print(f"RSCP Quantile (q_hat): {q_hat:.4f}")

# Evaluate RSCP on clean and backdoor inputs
clean_acc, clean_coverage, clean_set_size = evaluate_rscp_with_backdoor(
    model, test_loader, q_hat, sigma, num_samples, use_backdoor=False
)
backdoor_acc, backdoor_coverage, backdoor_set_size = evaluate_rscp_with_backdoor(
    model, test_loader, q_hat, sigma, num_samples, use_backdoor=True
)

print(f"RSCP Clean: Accuracy={clean_acc:.4f}, Coverage={clean_coverage:.4f}, Set Size={clean_set_size:.2f}")
print(f"RSCP Backdoor: Accuracy={backdoor_acc:.4f}, Coverage={backdoor_coverage:.4f}, Set Size={backdoor_set_size:.2f}")

# Plot clean vs backdoor coverage and set size
plot_rscp_metrics(
    (clean_coverage, clean_set_size),
    (backdoor_coverage, backdoor_set_size)
)

# Plot clean vs backdoor coverage and set size
def plot_rscp_metrics(clean_metrics, backdoor_metrics, labels=["Coverage", "Set Size"]):
    clean_coverage, clean_set_size = clean_metrics
    backdoor_coverage, backdoor_set_size = backdoor_metrics

    bar_width = 0.35
    x = np.arange(len(labels))

    plt.figure(figsize=(10, 6))
    plt.bar(x, [clean_coverage, clean_set_size], bar_width, label="Clean")
    plt.bar(x + bar_width, [backdoor_coverage, backdoor_set_size], bar_width, label="Backdoor")
    plt.xlabel("Metrics")
    plt.ylabel("Value")
    plt.xticks(x + bar_width / 2, labels)
    plt.title("RSCP Metrics: Clean vs Backdoor")
    plt.legend()
    plt.show()

plot_rscp_metrics(
    (clean_coverage, clean_set_size),
    (backdoor_coverage, backdoor_set_size)
)

# Visualize examples
def visualize_examples(data_loader, use_backdoor=False, num_samples=5):
    images, labels = next(iter(data_loader))
    if use_backdoor:
        images = add_backdoor_trigger(images)

    fig, axes = plt.subplots(1, num_samples, figsize=(15, 5))
    for i in range(num_samples):
        img = images[i].permute(1, 2, 0).cpu().numpy()
        img = (img * 0.5) + 0.5  # Unnormalize
        label = labels[i].item()
        axes[i].imshow(img)
        axes[i].axis("off")
        axes[i].set_title(f"Label: {label}")
    plt.show()

print("Clean Examples:")
visualize_examples(test_loader, use_backdoor=False)

print("Backdoor Examples:")
visualize_examples(test_loader, use_backdoor=True)

```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import numpy as np
import matplotlib.pyplot as plt

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data augmentation and normalization
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load CIFAR-10 dataset
train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)

# Define a simple trigger (white square in the bottom-right corner)
def add_backdoor_trigger(images, trigger_value=1.0):
    if images.ndimension() == 4:  # Batched images
        h, w = images.shape[-2:]
        trigger_size = min(5, h, w)  # Ensure trigger fits within the image
        images[:, :, -trigger_size:, -trigger_size:] = trigger_value
    elif images.ndimension() == 3:  # Single image
        h, w = images.shape[-2:]
        trigger_size = min(5, h, w)  # Ensure trigger fits within the image
        images[:, -trigger_size:, -trigger_size:] = trigger_value
    else:
        raise ValueError("Unsupported image dimensions. Expected 3D or 4D tensor.")
    return images

# Inject backdoor samples into the training dataset
def inject_backdoor(data_loader, target_class, poison_fraction=0.1):
    poisoned_images = []
    poisoned_labels = []
    clean_images = []
    clean_labels = []

    for images, labels in data_loader:
        for img, label in zip(images, labels):
            if np.random.rand() < poison_fraction:
                img = add_backdoor_trigger(img)
                poisoned_images.append(img)
                poisoned_labels.append(target_class)
            else:
                clean_images.append(img)
                clean_labels.append(label)

    all_images = torch.stack(clean_images + poisoned_images)
    all_labels = torch.tensor(clean_labels + poisoned_labels, dtype=torch.long)

    return DataLoader(torch.utils.data.TensorDataset(all_images, all_labels), batch_size=64, shuffle=True)

# Initialize a poisoned training loader with a target class
poisoned_train_loader = inject_backdoor(train_loader, target_class=0, poison_fraction=0.1)

# Define the model
model = resnet18(num_classes=10)
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train function
def train(model, loader, optimizer, criterion):
    model.train()
    running_loss = 0.0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    return running_loss / len(loader)

# Function to generate adversarial examples using FGSM
def generate_adversarial_examples(model, images, labels, epsilon=0.03):
    images = images.clone().detach().to(device)
    images.requires_grad = True

    outputs = model(images)
    loss = criterion(outputs, labels)
    model.zero_grad()
    loss.backward()

    perturbation = epsilon * images.grad.sign()
    adversarial_images = torch.clamp(images + perturbation, min=-1, max=1)  # Ensure valid range
    return adversarial_images

# Train function with adversarial examples
def train_with_adversarial_examples(model, loader, optimizer, criterion, epsilon=0.03):
    model.train()
    running_loss = 0.0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        # Generate adversarial examples
        adv_images = generate_adversarial_examples(model, images, labels, epsilon)

        # Combine clean and adversarial images
        combined_images = torch.cat([images, adv_images])
        combined_labels = torch.cat([labels, labels])

        # Forward pass
        outputs = model(combined_images)
        loss = criterion(outputs, combined_labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    return running_loss / len(loader)

# Evaluate function
def evaluate(model, loader, use_backdoor=False):
    model.eval()
    correct = 0
    total = 0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        if use_backdoor:
            images = add_backdoor_trigger(images)

        with torch.no_grad():
            outputs = model(images)
            _, predicted = outputs.max(1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    return correct / total

# Lists to store accuracies for plotting
clean_accuracies = []
adversarial_accuracies = []

# Training loop
epochs = 50
epsilon = 0.03  # Strength of adversarial perturbations

for epoch in range(epochs):
    train_loss = train_with_adversarial_examples(model, poisoned_train_loader, optimizer, criterion, epsilon=epsilon)

    # Evaluate clean accuracy
    clean_acc = evaluate(model, test_loader, use_backdoor=False)
    clean_accuracies.append(clean_acc)

    # Evaluate adversarial accuracy
    adversarial_acc = evaluate(model, test_loader, use_backdoor=False)
    adversarial_accuracies.append(adversarial_acc)

    print(f"Epoch {epoch + 1}/{epochs}, Loss: {train_loss:.4f}, Clean Accuracy: {clean_acc:.4f}, Adversarial Accuracy: {adversarial_acc:.4f}")

# Plot adversarial accuracy vs clean accuracy
plt.figure(figsize=(10, 6))
plt.plot(range(1, epochs + 1), clean_accuracies, label="Clean Accuracy", marker='o')
plt.plot(range(1, epochs + 1), adversarial_accuracies, label="Adversarial Accuracy", marker='x')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Adversarial Accuracy vs Clean Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import numpy as np
import matplotlib.pyplot as plt

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data augmentation and normalization
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load CIFAR-10 dataset
train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

train_size = int(0.8 * len(train_set))
calib_size = len(train_set) - train_size
train_set, calib_set = random_split(train_set, [train_size, calib_size])

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
calibration_loader = DataLoader(calib_set, batch_size=64, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)

# Define the model
model = resnet18(num_classes=10)
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Function to generate adversarial examples using FGSM
def generate_adversarial_examples(model, images, labels, epsilon=0.03):
    images = images.clone().detach().to(device)
    images.requires_grad = True

    outputs = model(images)
    loss = criterion(outputs, labels)
    model.zero_grad()
    loss.backward()

    perturbation = epsilon * images.grad.sign()
    adversarial_images = torch.clamp(images + perturbation, min=-1, max=1)  # Ensure valid range
    return adversarial_images

# Train function with adversarial examples
def train_with_adversarial_examples(model, loader, optimizer, criterion, epsilon=0.03):
    model.train()
    running_loss = 0.0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        # Generate adversarial examples
        adv_images = generate_adversarial_examples(model, images, labels, epsilon)

        # Combine clean and adversarial images
        combined_images = torch.cat([images, adv_images])
        combined_labels = torch.cat([labels, labels])

        # Forward pass
        outputs = model(combined_images)
        loss = criterion(outputs, combined_labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    return running_loss / len(loader)

# RSCP calibration function
def conformal_rscp(model, calibration_loader, sigma, num_samples, epsilon):
    model.eval()
    scores = []

    for images, _ in calibration_loader:
        images = images.to(device)

        # Generate scores for each sample
        with torch.no_grad():
            for _ in range(num_samples):
                noise = torch.normal(0, sigma, size=images.shape).to(device)
                noisy_images = images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                max_probs, _ = probs.max(dim=1)
                scores.extend((1 - max_probs).cpu().numpy())

    # Compute quantile for calibration
    q_hat = np.quantile(scores, 1 - epsilon)
    return q_hat

# Evaluate RSCP metrics
def evaluate_rscp_with_adversarial(model, loader, q_hat, sigma, num_samples, use_adversarial=False, epsilon=0.03):
    model.eval()
    total = 0
    correct = 0
    covered = 0
    total_set_size = 0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)
        total += labels.size(0)

        if use_adversarial:
            images = generate_adversarial_examples(model, images, labels, epsilon)

        with torch.no_grad():
            smooth_probs = []
            for _ in range(num_samples):
                noise = torch.normal(0, sigma, size=images.shape).to(device)
                noisy_images = images + noise
                outputs = model(noisy_images)
                probs = torch.softmax(outputs, dim=1)
                smooth_probs.append(probs)

            avg_probs = torch.stack(smooth_probs).mean(dim=0)
            prediction_sets = avg_probs >= q_hat

            for i, label in enumerate(labels):
                if prediction_sets[i, label].item():
                    covered += 1
                total_set_size += prediction_sets[i].sum().item()

            _, predicted = avg_probs.max(dim=1)
            correct += (predicted == labels).sum().item()

    accuracy = correct / total
    coverage = covered / total
    avg_set_size = total_set_size / total
    return accuracy, coverage, avg_set_size

# RSCP Parameters
sigma = 0.1
num_samples = 10
epsilon = 0.1

# Training loop
epochs = 40
train_epsilon = 0.03
clean_accuracies = []
adversarial_accuracies = []

for epoch in range(epochs):
    # Train with adversarial examples
    train_loss = train_with_adversarial_examples(model, train_loader, optimizer, criterion, epsilon=train_epsilon)

    # Calibrate RSCP
    q_hat = conformal_rscp(model, calibration_loader, sigma, num_samples, epsilon)
    print(f"Epoch {epoch + 1}/{epochs}, Loss: {train_loss:.4f}, RSCP Quantile (q_hat): {q_hat:.4f}")

    # Evaluate clean RSCP
    clean_acc, clean_coverage, clean_set_size = evaluate_rscp_with_adversarial(
        model, test_loader, q_hat, sigma, num_samples, use_adversarial=False, epsilon=train_epsilon
    )
    clean_accuracies.append(clean_acc)

    # Evaluate adversarial RSCP
    adv_acc, adv_coverage, adv_set_size = evaluate_rscp_with_adversarial(
        model, test_loader, q_hat, sigma, num_samples, use_adversarial=True, epsilon=train_epsilon
    )
    adversarial_accuracies.append(adv_acc)

    print(f"Clean Accuracy: {clean_acc:.4f}, Coverage: {clean_coverage:.4f}, Avg Set Size: {clean_set_size:.2f}")
    print(f"Adversarial Accuracy: {adv_acc:.4f}, Coverage: {adv_coverage:.4f}, Avg Set Size: {adv_set_size:.2f}")

# Plot clean vs adversarial accuracies
plt.figure(figsize=(10, 6))
plt.plot(range(1, epochs + 1), clean_accuracies, label="Clean Accuracy", marker='o')
plt.plot(range(1, epochs + 1), adversarial_accuracies, label="Adversarial Accuracy", marker='x')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Clean vs Adversarial Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```
